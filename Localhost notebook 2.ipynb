```python
import os
import cv2
import numpy as np
import sklearn.model_selection
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
os.chdir('c:\\users\\timmo\\onedrive\\ITU\\5g computing challenge aws verizon\\pictures 01062021')
picture_names = os.listdir()
print(len(picture_names))
```

```python
labels = []
for picture in picture_names:
    label = picture.split()[1].split('.')[0]
    labels.append(label.replace('n', 'N'))

for i in range(len(picture_names)):
    print(picture_names[i] + ' ' + labels[i])
```

```python
print(len(labels))
print(labels.count('N'))
print(labels.count('G'))
print(list(set(labels)))
```

```python
list_of_arrays = []
for picture in picture_names:
    pic = cv2.imread(picture)
    pic = cv2.resize(pic, (300, 300))
    pic = pic.flatten()
    pic = np.asarray(pic)
    list_of_arrays.append(pic)
array_of_arrays = np.asarray(list_of_arrays)
```

```python
print(labels.count('N'))
print(labels.count('G'))
print(array_of_arrays.shape)
array_of_labels = np.asarray(labels)
print(array_of_labels.shape)
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(array_of_arrays, array_of_labels, test_size=0.2, random_state = 0)
VAR_rfc = RandomForestClassifier().fit(X_train, y_train)
y_pred = VAR_rfc.predict(X_test)
print(sklearn.metrics.classification_report(y_test, y_pred))
```
