```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pickle
from sklearn.ensemble import RandomForestClassifier
import cv2
import numpy as np
```

```python
picture_name = '20210601_194527 G.jpg'
pic_show = mpimg.imread(picture_name)
plt.imshow(pic_show)
```

```python
a = open('file_from_nb', 'rb')
unpickled_model = pickle.load(a)
```

```python
pic_flat = cv2.imread(picture_name)
pic_flat = cv2.resize(pic_flat, (300,300))
pic_flat = pic_flat.flatten()
pic_flat = np.asarray(pic_flat)
pic_flat = pic_flat.reshape(1, -1)
prediction = unpickled_model.predict(pic_flat)
prediction = str(prediction)
for k in ['[', ']', "'",]:
    prediction = prediction.replace(k, '')
print(prediction)
print('Picture name is ' + picture_name)
print('Prediction is ' + prediction)
pic_show = mpimg.imread(picture_name)
plt.imshow(pic_show)
```
